```python
def indicator_fill(df):
    a = ta.trend.ema_indicator(close=df.Close,window=5)
    f = ta.trend.sma_indicator(close=df.Close,window=50)
    i = ta.trend.sma_indicator(close=df.Close,window=20)
    c = ta.volatility.average_true_range(high=df.High,low=df.Low,close=df.Close)
    ch = ta.volume.chaikin_money_flow(high=df.High,low=df.Low,close=df.Close,volume=df.Volume,window=20)
    dp = ta.trend.dpo(close = df.Close,window=20)
    pv = ta.momentum.pvo(volume=df.Volume, window_slow = 20, window_fast = 5)
    rs = ta.momentum.rsi(close=df.Close, window=14)
    df["5_EMA"] = a
    df["20_SMA"] = i
    df['ATR'] = c
    df['Chaikin'] = ch
    df['dpo'] = dp
    df['pvo'] = pv
    df['rsi'] = rs
    return df
```

```python
import math
import statistics
def buy_quality(hi,lo,clo,start,bl):
    if bl =='buy' and abs(float(max(hi[1:])) - float(start)) > abs(float(min(lo[1:])) - float(start)):
        return "go buy"
    else:
        return "wrong go buy"

def sell_quality(hi,lo,clo,start,sl):
    if sl == 'sell' and abs(float(min(lo[1:])) - float(start)) > abs(float(max(hi[1:])) - float(start)):
        return "go sell"
    else:
        return "wrong go sell"

def mover(hi,lo,clo,start):
    up_move = max(hi[1:]) - start
    low_move = start - min(lo[1:])
    return up_move, low_move

def buy_calc_side(hi,lo,clo,start):
    if float(statistics.mean(clo[1:])) > float(start):
        return "buy"
    else:
        return "wrong buy"

def sell_calc_side(hi,lo,clo,start):
    if float(statistics.mean(clo[1:])) < float(start):
        return "sell"
    else:
        return "wrong sell"
    
def label_calc(df):
    buy_lab = []
    sell_lab = []
    bq = []
    sq = []
    go = []
    profit = []
    loss = []
    k = 30
    for t in range(0,len(df)):
        if t >= len(df) - k:
            break
        a1 = df[t:t+k]
        a0 = df[t:t+k]
        st = a0.Close.tolist()[0]
        hi = a1.High.tolist()
        lo = a1.Low.tolist()
        clo = a1.Close.tolist()
        buy_label = buy_calc_side(hi,lo,clo,st)
        sell_label = sell_calc_side(hi,lo,clo,st)
        b_quality = buy_quality(hi,lo,clo,st,buy_label)
        s_quality = sell_quality(hi,lo,clo,st,sell_label)
        #up_move, low_move = mover(hi,lo,clo,st)
        
        
        buy_lab += [buy_label]
        sell_lab += [sell_label]
        bq += [b_quality]
        sq += [s_quality]
        
        
        #profit += [up_move]
        #loss += [low_move]

    if len(buy_lab) < len(df):
        buy_lab += ["Pass"] * (len(df)-len(buy_lab))
    if len(sell_lab) < len(df):
        sell_lab += ["Pass"] * (len(df)-len(sell_lab))
    if len(sq) < len(df):
        sq += ["Pass"] * (len(df)-len(sq))
    if len(bq) < len(df):
        bq += ["Pass"] * (len(df)-len(bq))

    df["buy_label"] = buy_lab
    df["sell_label"] = sell_lab
    df["bu_quality"] = bq
    df["se_quality"] = sq
    

    return df

def label_calc_v2(df):
    buy_lab = []
    sell_lab = []
    bq = []
    sq = []
    go = []
    profit = []
    loss = []
    k = 30
    for t in range(0,len(df)):
        if t >= len(df) - k:
            break
        a1 = df[t:t+k]
        a0 = df[t:t+k]
        st = a0.Close.tolist()[0]
        hi = a1.High.tolist()
        lo = a1.Low.tolist()
        clo = a1.Close.tolist()
        buy_label = buy_calc_side(hi,lo,clo,st)
        sell_label = sell_calc_side(hi,lo,clo,st)
        b_quality = buy_quality(hi,lo,clo,st,buy_label)
        s_quality = sell_quality(hi,lo,clo,st,sell_label)
        #up_move, low_move = mover(hi,lo,clo,st)
        
        
        buy_lab += [buy_label]
        sell_lab += [sell_label]
        bq += [b_quality]
        sq += [s_quality]
        
        
        #profit += [up_move]
        #loss += [low_move]

    if len(buy_lab) < len(df):
        buy_lab += ["Pass"] * (len(df)-len(buy_lab))
    if len(sell_lab) < len(df):
        sell_lab += ["Pass"] * (len(df)-len(sell_lab))
    if len(sq) < len(df):
        sq += ["Pass"] * (len(df)-len(sq))
    if len(bq) < len(df):
        bq += ["Pass"] * (len(df)-len(bq))

    df["buy_label_v2"] = buy_lab
    df["sell_label_v2"] = sell_lab
    df["bu_quality_v2"] = bq
    df["se_quality_v2"] = sq
    

    return df

def label_calc_v3(df):
    buy_lab = []
    sell_lab = []
    bq = []
    sq = []
    go = []
    profit = []
    loss = []
    k = 2
    for t in range(0,len(df)):
        if t >= len(df) - k:
            break
        a1 = df[t:t+k]
        clo = a1.Close.tolist()
        #up_move, low_move = mover(hi,lo,clo,st)
        
        
        buy_lab += [clo[-1]]
        
        
        #profit += [up_move]
        #loss += [low_move]

    if len(buy_lab) < len(df):
        buy_lab += ["Pass"] * (len(df)-len(buy_lab))


    df["next_value"] = buy_lab

    

    return df
```

```python
from sklearn.preprocessing import normalize
import math
def df_vectorize(df):
    vec_load = []
    m5_vec = []
    buy_side = []
    sell_side = []
    b_quality = []
    s_quality = []
    take_profit = []
    stop_loss = []
    print(len(df))
    for h in range(100,len(df)):
        buff_df = df[h-100:h]
        if buff_df.buy_label.tolist()[-1] != "Skip":
            
            orig_close = buff_df.Close.tolist()
            orig_hi = buff_df.High.tolist()
            orig_lo = buff_df.Low.tolist()
            orig_op = buff_df.Open.tolist()
            orig_volume  = buff_df.Volume.tolist()
            us_vec = buff_df.us.tolist()
            ls_vec = buff_df.ls.tolist()
            rb_vec = buff_df.rb.tolist()
            ma_diff = buff_df.ma_diff.tolist()
            atr = buff_df.ATR.tolist()
            chaikin = buff_df.Chaikin.tolist()
            dpo = buff_df.dpo.tolist()
            pvo = buff_df.pvo.tolist()
            rsi = buff_df.rsi.tolist()
            
            b_quality += [buff_df.bu_quality.tolist()[-1]]
            buy_side += [buff_df.buy_label.tolist()[-1]]
            
            chaikin = [0 if math.isnan(x) or math.isinf(x) else x for x in chaikin]
            chaikin = list(normalize(np.array(chaikin).reshape(1,-1), axis=1).flatten())
            ma_diff = list(normalize(np.array(ma_diff).reshape(1,-1), axis=1).flatten())
            orig_volume = list(normalize(np.array(orig_volume).reshape(1,-1), axis=1).flatten())
            rb_vec = list(normalize(np.array(rb_vec).reshape(1,-1), axis=1).flatten())
            atr = list(normalize(np.array(atr).reshape(1,-1), axis=1).flatten())
            dpo = list(normalize(np.array(dpo).reshape(1,-1), axis=1).flatten())
            pvo = list(normalize(np.array(pvo).reshape(1,-1), axis=1).flatten())
            rsi = list(normalize(np.array(rsi).reshape(1,-1), axis=1).flatten())
            
            #vec_load += [close_log + rb_vec + us_vec + ls_vec + ma_diff + tclose_diff + fclose_diff]
            #m5_vec += [orig_close + orig_hi + orig_lo + orig_op + ma_diff]
            #m5_vec += [us_vec + ls_vec + rb_vec + orig_close]
            #m5_vec += [chaikin + ma_diff + orig_volume + rb_vec]
            #m5_vec += [ma_diff + pvo + dpo + chaikin + rsi + atr]
            m5_vec += [ma_diff + chaikin]
        else:
            print('Skip')
    print('Vector_Done')
    vec_load = m5_vec
    return vec_load, buy_side, b_quality
        
def df_vectorize_v2(df):
    vec_load = []
    m5_vec = []
    buy_side = []
    sell_side = []
    b_quality = []
    s_quality = []
    take_profit = []
    stop_loss = []
    print(len(df))
    for h in range(100,len(df)):
        buff_df = df[h-100:h]
        if buff_df.buy_label.tolist()[-1] != "Skip":
            
            orig_close = buff_df.Close.tolist()

            buy_side += [buff_df.next_value.tolist()[-1]]

            m5_vec += [orig_close + chaikin]
        else:
            print('Skip')
    print('Vector_Done')
    vec_load = m5_vec
    return vec_load, buy_side
```
